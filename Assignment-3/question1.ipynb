```python
# Lab Assignment - 3
# DIP Lab
# Date - 27 Aug 19
# Name - Ashish Goyal
# Id - 2016ucp1100
import cv2  
import numpy as np
import math 
#function cv2.imread() to read an image
img_object = cv2.imread('house.jpg',0) #Loads a color image. Any transparency of image will be neglected. It is the default flag.
print(type(img_object))
#print(img_object1) #to show image in matrix form
height = img_object.shape[0]
width = img_object.shape[1]
cv2.imshow('Initial Image',img_object)
print(height,width)
#print(img_object)

laplacian_filter1 = [[1,1,1],[1,-8,1],[1,1,1]]
laplacian_filter2 = [[-1,2,-1],[2,-4,2],[-1,2,-1]]


final_img1 = cv2.imread('house.jpg',0)
final_img2 = cv2.imread('house.jpg',0)
final_img3 = cv2.imread('house.jpg',0)
final_img4 = cv2.imread('house.jpg',0)
final_img5 = cv2.imread('house.jpg',0)
final_img6 = cv2.imread('house.jpg',0)
final_img7 = cv2.imread('house.jpg',0)
final_img8 = cv2.imread('house.jpg',0)
final_img9 = cv2.imread('house.jpg',0)
#final_img =final_img[1:height, 1:width]

def gausian(sigma):
    list1=[]
    for i in range(-1,2):
        list2=[]             
        for j in range(-1,2):
            t=(math.e**(-1*(i*i+j*j)/(2*sigma*sigma)))/(2*math.pi*sigma*sigma)
            list2.append(t)
        list1.append(list2)
    return list1
                     
def filter(input_img,final_img,height,width,filter_matrix):
    for i in range(1,height-1):
        for j in range (1,width-1):
            val=0
            for x in range(-1,2):
                for y in range(-1,2):
                    val= val + (filter_matrix[x+1][y+1]*input_img[i+x][j+y])
            final_img[i][j] = val
    return final_img
            
gausian_1 = gausian(1)
gausian_2 = gausian(2)
gausian_3 = gausian(3)


final_img1 = filter(img_object,final_img1,height,width,gausian_1)
cv2.imshow('Gausian Image sigma=1',final_img1)
cv2.imwrite('output_Gausian_Image_sigma1.jpg',final_img1)

final_img2 = filter(img_object,final_img2,height,width,gausian_2)
cv2.imshow('Gausian Image sigma=2',final_img2)
cv2.imwrite('output_Gausian_Image_sigma2.jpg',final_img2)

final_img3 = filter(img_object,final_img3,height,width,gausian_3)
cv2.imshow('Gausian Image sigma=3',final_img3)
cv2.imwrite('output_Gausian_Image_sigma3.jpg',final_img3)

final_img4 = filter(final_img1,final_img4,height,width,laplacian_filter1)
cv2.imshow('output_Laplacian_Image1_gausian1',final_img4)
cv2.imwrite('output_Laplacian_Image1_gausian1.jpg',final_img4)

final_img5 = filter(final_img1,final_img5,height,width,laplacian_filter2)
cv2.imshow('output_Laplacian_Image2_gausian1',final_img4)
cv2.imwrite('output_Laplacian_Image2_gausian1.jpg',final_img5)

final_img6 = filter(final_img2,final_img6,height,width,laplacian_filter1)
cv2.imshow('output_Laplacian_Image1_gausian2',final_img6)
cv2.imwrite('output_Laplacian_Image1_gausian2.jpg',final_img6)

final_img7 = filter(final_img2,final_img7,height,width,laplacian_filter2)
cv2.imshow('output_Laplacian_Image2_gausian2',final_img7)
cv2.imwrite('output_Laplacian_Image2_gausian2.jpg',final_img7)

final_img8 = filter(final_img3,final_img8,height,width,laplacian_filter1)
cv2.imshow('output_Laplacian_Image1_gausian3',final_img8)
cv2.imwrite('output_Laplacian_Image1_gausian3.jpg',final_img8)

final_img9 = filter(final_img3,final_img9,height,width,laplacian_filter2)
cv2.imshow('output_Laplacian_Image2_gausian3',final_img9)
cv2.imwrite('output_Laplacian_Image2_gausian3.jpg',final_img9)


k = cv2.waitKey(0) & 0xFF    #for 64 bit machine
cv2.destroyAllWindows()
```
