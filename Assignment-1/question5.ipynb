```python
# Lab Assignment - 1
# DIP Lab
# Date - 6 Aug 19
# Name - Ashish Goyal
# Id - 2016ucp1100
import cv2
import matplotlib.pyplot as plt

def histogram(height, width, gray):
    frequency_list = [0] * 256
    for i in range(0, height):
        for j in range(0, width):
            intensity = gray[i][j]
            frequency_list[intensity] += 1

    plt.bar([i for i in range(0, 256)], frequency_list, width=0.30)
    plt.xlabel("Intensity of pixel ")
    plt.ylabel("frequency")
    plt.title("Histogram")
    plt.show()
    



img_object = cv2.imread('img.png')
gray = cv2.cvtColor(img_object, cv2.COLOR_BGR2GRAY)
height, width = gray.shape
histogram(height, width, gray)
```
